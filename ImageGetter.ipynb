```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def get_anime_image(category='sfw/waifu'):
    """
    Mengambil URL gambar anime dari API waifu.pics.

    Args:
        category (str): Kategori gambar yang diinginkan.
                        Contoh: 'sfw/waifu', 'sfw/neko', 'sfw/shinobu',
                                'nsfw/waifu', 'nsfw/neko', 'nsfw/trap', 'nsfw/blowjob'
                                (gunakan kategori NSFW dengan hati-hati dan sesuai konteks)

    Returns:
        str: URL gambar anime, atau None jika terjadi error.
    """
    base_url = "https://api.waifu.pics/"
    full_url = f"{base_url}{category}"
    try:
        response = requests.get(full_url)
        response.raise_for_status()  # Akan memunculkan HTTPError untuk status kode error
        data = response.json()
        return data.get('url')
    except requests.exceptions.RequestException as e:
        print(f"Terjadi kesalahan saat mengambil data dari API: {e}")
        return None

def display_image_from_url(image_url):
    """
    Menampilkan gambar dari URL menggunakan Pillow dan Matplotlib.

    Args:
        image_url (str): URL gambar.
    """
    if not image_url:
        print("Tidak ada URL gambar yang diberikan.")
        return

    try:
        response = requests.get(image_url)
        response.raise_for_status() # Akan memunculkan HTTPError untuk status kode error
        image_bytes = BytesIO(response.content)
        img = Image.open(image_bytes)

        plt.imshow(img)
        plt.axis('off') # Sembunyikan sumbu
        plt.show()
    except requests.exceptions.RequestException as e:
        print(f"Terjadi kesalahan saat mengunduh gambar: {e}")
    except IOError as e:
        print(f"Tidak dapat membuka atau menampilkan gambar: {e}")

# --- Cara Penggunaan ---
if __name__ == "__main__":
    print("Memulai pengambilan gambar anime...")

    # Contoh penggunaan SFW (Safe For Work)
    # Anda bisa mencoba kategori lain seperti 'sfw/neko', 'sfw/shinobu'
    image_url_sfw = get_anime_image(category='sfw/waifu')
    if image_url_sfw:
        print(f"URL Gambar SFW Waifu: {image_url_sfw}")
        display_image_from_url(image_url_sfw)
    else:
        print("Gagal mendapatkan gambar SFW.")

    print("\nMencoba mengambil gambar kategori lain (sfw/neko)...")
    image_url_neko = get_anime_image(category='sfw/neko')
    if image_url_neko:
        print(f"URL Gambar SFW Neko: {image_url_neko}")
        display_image_from_url(image_url_neko)
    else:
        print("Gagal mendapatkan gambar Neko.")

    # Peringatan: Gunakan kategori NSFW dengan bijak dan hanya di lingkungan yang sesuai.
    # Harap pastikan Anda memahami kontennya sebelum menjalankannya.
    # Untuk menjalankan contoh NSFW, hilangkan komentar di baris di bawah ini:
    # print("\nMemperingatkan: Akan mencoba mengambil gambar NSFW (nsfw/waifu).")
    # print("Pastikan Anda nyaman dengan konten ini sebelum melanjutkan.")
    # image_url_nsfw = get_anime_image(category='nsfw/neko')
    # if image_url_nsfw:
    #     print(f"URL Gambar NSFW Waifu: {image_url_nsfw}")
    #     display_image_from_url(image_url_nsfw)
    # else:
    #     print("Gagal mendapatkan gambar NSFW.")
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def get_anime_image_nekoslife(endpoint_type='waifu'):
    """
    Mengambil URL gambar anime dari API nekos.life berdasarkan tipe/endpoint.

    Args:
        endpoint_type (str): Tipe gambar yang diinginkan.
                             Contoh SFW: 'waifu', 'neko', 'kitsune', 'hug', 'kiss', 'pat', 'smug', 'baka', 'poke', 'slap', 'cuddle', 'feed', 'cuddle', 'tickle', 'fox_girl', 'kemonomimi'
                             Contoh NSFW: 'ero', 'lewd', 'feet', 'cum', 'trap', 'futanari', 'ero_neko', 'solo', 'yuri', 'blowjob', 'pussy', 'tits', 'anal', 'hentai'
                             (Gunakan kategori NSFW dengan sangat hati-hati dan sesuai konteks!)

    Returns:
        str: URL gambar anime, atau None jika terjadi error.
    """
    base_url = "https://nekos.life/api/v2/img/"
    full_url = f"{base_url}{endpoint_type}"
    try:
        response = requests.get(full_url)
        response.raise_for_status()  # Akan memunculkan HTTPError untuk status kode error
        data = response.json()
        return data.get('url')
    except requests.exceptions.RequestException as e:
        print(f"Terjadi kesalahan saat mengambil data dari API Nekos.life: {e}")
        return None

def display_image_from_url(image_url):
    """
    Menampilkan gambar dari URL menggunakan Pillow dan Matplotlib.

    Args:
        image_url (str): URL gambar.
    """
    if not image_url:
        print("Tidak ada URL gambar yang diberikan.")
        return

    try:
        response = requests.get(image_url)
        response.raise_for_status() # Akan memunculkan HTTPError untuk status kode error
        image_bytes = BytesIO(response.content)
        img = Image.open(image_bytes)

        plt.imshow(img)
        plt.axis('off') # Sembunyikan sumbu
        plt.show()
    except requests.exceptions.RequestException as e:
        print(f"Terjadi kesalahan saat mengunduh gambar: {e}")
    except IOError as e:
        print(f"Tidak dapat membuka atau menampilkan gambar: {e}")

# --- Cara Penggunaan ---
if __name__ == "__main__":
    print("Memulai pencarian gambar anime dengan Nekos.life API...")

    # Contoh penggunaan SFW (Safe For Work)
    # Masukkan 'tipe' atau 'endpoint' yang ingin Anda cari
    search_query_sfw = input("Masukkan tipe gambar SFW yang ingin Anda cari (misal: waifu, neko, kiss, hug): ").lower()
    image_url_sfw = get_anime_image_nekoslife(endpoint_type=search_query_sfw)
    if image_url_sfw:
        print(f"URL Gambar SFW '{search_query_sfw}': {image_url_sfw}")
        display_image_from_url(image_url_sfw)
    else:
        print(f"Gagal mendapatkan gambar untuk tipe '{search_query_sfw}'. Pastikan tipenya benar.")

    # Peringatan: Gunakan kategori NSFW dengan bijak dan hanya di lingkungan yang sesuai.
    # Harap pastikan Anda memahami kontennya sebelum menjalankan.
    # Untuk menjalankan contoh NSFW, hilangkan komentar di baris di bawah ini:
    #
    print("\n--- Peringatan: Akan mencoba mengambil gambar NSFW ---")
    print("Pastikan Anda nyaman dengan konten ini sebelum melanjutkan.")
    search_query_nsfw = input("Masukkan tipe gambar NSFW yang ingin Anda cari (misal: hentai, trap, blowjob): ").lower()
    image_url_nsfw = get_anime_image_nekoslife(endpoint_type=search_query_nsfw)
    if image_url_nsfw:
        print(f"URL Gambar NSFW '{search_query_nsfw}': {image_url_nsfw}")
        display_image_from_url(image_url_nsfw)
    else:
        print(f"Gagal mendapatkan gambar untuk tipe NSFW '{search_query_nsfw}'. Pastikan tipenya benar.")
```
